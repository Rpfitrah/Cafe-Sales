```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Buat DataFrame dari data di atas
data = pd.DataFrame({
    'social_media': [200, 230, 180, 300, 250, 280, 350, 270, 310, 220],
    'search_engine': [100, 120, 90, 200, 150, 180, 250, 160, 190, 110],
    'sales': [1000, 1200, 900, 1500, 1300, 1450, 1800, 1350, 1520, 1100]
})

# 2. Hitung Matrix Korelasi
correlation_matrix = data.corr()
print("MATRIX KORELASI:")
print(correlation_matrix)
print("\n")

# 3. Visualisasi dengan Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Heatmap Korelasi Budget Iklan vs Penjualan')
plt.show()

# 4. Visualisasi dengan Scatter Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter Plot Social Media vs Sales
sns.scatterplot(x=data['social_media'], y=data['sales'], ax=axes[0])
axes[0].set_title('Social Media Budget vs Sales')
axes[0].set_xlabel('Budget Social Media')
axes[0].set_ylabel('Penjualan')

# Scatter Plot Search Engine vs Sales
sns.scatterplot(x=data['search_engine'], y=data['sales'], ax=axes[1])
axes[1].set_title('Search Engine Budget vs Sales')
axes[1].set_xlabel('Budget Search Engine')
axes[1].set_ylabel('Penjualan')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import pearsonr

# Uji signifikansi korelasi untuk Social Media
correlation_social, p_value_social = pearsonr(data['social_media'], data['sales'])

print(f"Korelasi Social Media vs Sales: {correlation_social:.4f}")
print(f"P-Value: {p_value_social:.4f}")

# Keputusan Statistik
alpha = 0.05
if p_value_social < alpha:
    print("Keputusan: Tolak H0. Hubungan antara Social Media Budget dan Penjualan SIGNIFIKAN secara statistik.")
else:
    print("Keputusan: Gagal Tolak H0. Tidak cukup bukti untuk mengatakan hubungannya signifikan.")
```

**Uji Korelasi: Social Media Budget vs Sales**

- **Koefisien Korelasi**: 0.9941 → Hubungan sangat kuat dan positif
- **P-Value**: 0.0000
- **Alpha**: 0.05

**Hipotesis:**
- H₀: Tidak ada hubungan linear antara anggaran social media dan penjualan
- H₁: Ada hubungan linear yang signifikan antara anggaran social media dan penjualan

**Keputusan Statistik:** Karena p-value < alpha, maka H₀ ditolak. Terdapat hubungan yang signifikan antara anggaran social media dan penjualan.

**Interpretasi Bisnis:** Anggaran iklan di media sosial memiliki dampak yang sangat kuat terhadap peningkatan penjualan. Ini bisa menjadi dasar untuk mempertimbangkan peningkatan alokasi budget di kanal tersebut.
